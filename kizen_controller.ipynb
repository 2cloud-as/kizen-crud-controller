```python
import requests
import json
from dotenv import load_dotenv
import os
```

```python
# Load environment variables from .env file
load_dotenv()
```

```python
base_url = "https://app.fmo.kizen.com/api"

url = f"{base_url}/client/v2"
api_key = os.getenv('KIZEN_API_KEY')
user_id = os.getenv('KIZEN_USER_ID')
business_id = os.getenv('KIZEN_BUSINESS_ID')

payload={}
headers = {
    'X-API-Key': api_key,
    'X-User-Id': user_id,
    'X-Business-Id': business_id,
    'CONTENT-TYPE': 'application/json'
}
```

```python
names = []
params = {
     'page_size': 50,
     'page': 1
 }
```

```python
response = requests.request("POST", url, headers=headers, data=payload)
```

```python
# Check the response
if response.status_code == 200:
    print("Request successful!")
    print(response.json())  # Print the response JSON
else:
    print(f"Request failed with status code: {response.status_code}")
    print(response.text)  # Print the error message
```
